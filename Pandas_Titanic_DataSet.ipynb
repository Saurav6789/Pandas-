```python
#count the no of null values in the column

train_df.isna().sum()
```

```python
#Counts the values for  each type in Embarked column

train_df["Embarked"].value_counts()

```

```python
# We have S most occuring so will replace null value with S

train_df["Embarked"]=train_df["Embarked"].fillna("S")
```

```python
#Cross Checking the value of S count increases
train_df["Embarked"].value_counts()
```

```python
#Replacing the Age column null values with the median 

train_df["Age"].fillna(train_df['Age'].median(), inplace = True)
```

```python
#Check that the null values are replaced or not 

train_df["Age"].isna().any()
```

```python
# Calculate the no of males that survived and no of males not survived 

print(train_df["Survived"][train_df["Sex"] == 'male'].value_counts())
print(train_df["Survived"][train_df["Sex"]=='female'].value_counts())
```

```python
#Normalize the male and female survival
print(train_df["Survived"][train_df["Sex"] == 'male'].value_counts(normalize=True))
print(train_df["Survived"][train_df["Sex"] == 'female'].value_counts(normalize=True))
```

It looks like from above that the chances of female survival is more than males

```python
#Checking the Age of the person is related to the survival 

#Calculatig the avaerage age of the male and female who survived 
survived = train_df['Survived'] == 1
male = train_df['Sex'] == 'male'
female = train_df['Sex'] == 'female'

print('Average age of male who survived', train_df[survived & male].Age.mean())
print('Average age of female who survived', train_df[survived & female].Age.mean())
```

```python
#Check how many old or child (below 18) survived

#Create a child column and assig NaN

train_df["Child"] = float('NaN')

# Assign 1 to passengers under 18, 0 to those 18 or older. Print the new column

train_df["Child"][train_df["Age"] < 18] = 1
train_df["Child"][train_df["Age"] >= 18] = 0

#print the no of survival who are adults 
print(train_df["Survived"][train_df["Child"] == 0].value_counts())

#Print the no of survival who are child 

print(train_df["Survived"][train_df["Child"] == 1].value_counts())

# Print normalized Survival Rates for passengers 18 or older
print(train_df["Survived"][train_df["Child"] == 0].value_counts(normalize = True))

# Print normalized Survival Rates for passengers under 18
print(train_df["Survived"][train_df["Child"] == 1].value_counts(normalize = True))
```
